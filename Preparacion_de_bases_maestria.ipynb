#Quitar acentos

```python
def remover_tildes(texto):
    """
    Función que elimina los acentos de un texto.

    Argumentos:
      texto (str): El texto al que se le quieren eliminar los acentos.

    Retorna:
      str: El texto sin acentos.
    """
    diccionario = {'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
                   'Á': 'A', 'É': 'E', 'Í': 'I', 'Ó': 'O', 'Ú': 'U'}
    # Crear la tabla de traducción usando str.maketrans()
    tabla_traduccion = str.maketrans(diccionario)
    # Usar la tabla para traducir el texto
    texto_sin_acentos = texto.translate(tabla_traduccion)
    return texto_sin_acentos
```
